```python
import math, matplotlib, time
from matplotlib import pyplot as plt
import numpy as np

st = time.monotonic()

gravity = 9.81
massPerson = 80
massMotorcycle = 170
massEngine = 45
massElectronics = 40
mass = massPerson + massMotorcycle - massEngine + massElectronics
power = 20e3
powers = [power, ]
voltage = 120
voltages = [voltage]
resistance = voltage ** 2 / power     #P = V^2/R
current = voltage/resistance
currents = [current, ]
#capacitors
capacitorsInSeries = math.ceil(48/2.7)
capacitorsParallel = 1
capacitors = capacitorsInSeries * capacitorsParallel
capacitorsVolume = capacitors * 30*30*70 / (25.4**3)
capacitance = (( (1/500)*capacitorsInSeries ) ** -1) * capacitorsParallel  # 1/Ct = 1/C1 + 1/C2 ..... 1/Cn

charge = capacitance * voltage #(capacitance)
charges = [charge,]
capacitorEnergy = (voltage ** 2) * 0.5 * capacitance   # 1/2 * C * V^2
capacitorEnergies = [capacitorEnergy, ]

electricalEfficiency = 0.9
mechanicalEfficiency = 0.9
kineticEnergy = 0
kineticEnergies = [kineticEnergy, ]
workdone = 0
workdones = [workdone]
generatorPower = 1500
generatorPowers = [generatorPower]
generatorEnergy = 0
generatorEnergies = [generatorEnergy]

displacement = 0
displacements = [displacement, ]
velocity = 0
velocities = [velocity, ]
kmhrs = []
acceleration = 0
accelerations = [acceleration]

force = 0
forces = [force]
#rolling resistance
coefficientOfRollingResistance = 0.006
wheelRadius = 19 * 0.0254 / 2
motionResistance = 0
motionResistances = [motionResistance]
#drag
dragCoefficient = 1.8
crossSectionalArea = 0.9
airDensity = 1.2
drag = 0
drags = [drag]
friction = 0
frictions = [friction]
forcesLoss = 0
forcesLosses = [forcesLoss]
mechanicalPowerLoss = 0
mechanicalPowerLosses = [mechanicalPowerLoss]

dt = 0.1
tMax = 200
t = 0
ts = [t]

def timeItOccured(val, listy):
    try:
        index = listy.index(val)
        timeAt = ts[index]
        return timeAt
    except ValueError:
        return None
def timeItOccuredSequential(val, listy, reverse=False):
    #get index
    for i, v in enumerate(listy):
        if reverse:
            if val >= v:
                index = i
                break
        else:
            if val <= v:
                index = i
                break
    #get corresponding value
    val = listy[index]
    return val
def occuredAtTime(time, listy):
    for i,tim in enumerate(ts):
        if time>tim:
            index = i
            break
    else:
        if len(listy) > 0:
            return listy[-1]
        else:
            return None
    #now, return value at time
    val = listy[index]
    return val
    
def calcAcceleration():
    #accel = power/(mass*velocity) #P = m*a*v
    accel = (velocity-velocities[-1])/dt
    return accel

def update():
    powers.append(power)
    voltages.append(voltage)
    currents.append(current)
    charges.append(charge)
    capacitorEnergies.append(capacitorEnergy)
    
    kineticEnergies.append(kineticEnergy)
    workdones.append(workdone)
    generatorPowers.append(generatorPower)
    generatorEnergies.append(generatorEnergy)
    
    displacements.append(displacement)
    velocities.append(velocity)
    accelerations.append(acceleration)
    
    frictions.append(friction)
    drags.append(drag)
    motionResistances.append(motionResistance)
    forcesLosses.append(forcesLoss)
    mechanicalPowerLosses.append(mechanicalPowerLoss)
    forces.append(force)
    
    ts.append(t)
    
def plotThem():
    #https://jakevdp.github.io/PythonDataScienceHandbook/04.12-three-dimensional-plotting.html
    matplotlib.rcParams['figure.figsize'] = [20, 5]
    figure, axes = plt.subplots(nrows=3, ncols=6)

    axes[0, 0].plot(ts, displacements)
    axes[0, 1].plot(ts, velocities)
    axes[0, 2].plot(ts, accelerations)
    axes[0, 4].plot(ts, kmhrs)
    
    axes[1, 0].plot(ts, voltages)
    axes[1, 1].plot(ts, currents)
    axes[1, 2].plot(ts, powers)
    axes[1, 2].plot(ts, mechanicalPowerLosses)
    axes[1, 2].plot(ts, generatorPowers)
    axes[1, 3].plot(ts, workdones)
    axes[1, 4].plot(ts, charges)
    axes[1, 5].plot(ts, capacitorEnergies)
    axes[1, 5].plot(ts, generatorEnergies)
    
    axes[2, 0].plot(ts, forces)
    axes[2, 0].plot(ts, forcesLosses)
    axes[2, 2].plot(ts, drags)
    axes[2, 4].plot(ts, motionResistances)
    axes[2, 5].plot(ts, frictions)

    #figure.tight_layout()

def importantInfo():
    print('''Note Well:
    These are energy transfer calculations.
    You should try force calculations (torque>force>accel>etc)
    Even with an IC engine generator, you will need high voltage to get high power generation (higher current charge).
    ''')
    print(f"""Total mass: {mass}
    Capacitance:{capacitance}, "Number of capacitors:{capacitors}, "Capacitors' volume:"{capacitorsVolume},
    Max accel:{max(accelerations)}'m/s^2', g's:{max(accelerations)/gravity}, max velocity:{max(velocities)*3.6}km/hr,
    Max mechanical power loss:{max(mechanicalPowerLosses)}
    """)
    print(f'''Limitations: 
    Not high enough power generating motors found(50), 
    5hp motor charges too slow(30sec to 5mins), 
    Capacitor depletes kinda fast (voltage drop too fast). Leads to massive power drop.
    System has to upscale to overcome mechanical power losses.
    Top speed about 50km/hr.
    The super capacitor pack will be huge {capacitorsVolume}inch^3.
    ''')

while(t<tMax):
    #body
    #power and energy
    workdone = power*dt
    capacitorEnergy -= workdone / electricalEfficiency
    kineticEnergy += workdone * mechanicalEfficiency    #ME to be replaced by friction
    
    #voltage, current
    voltage = (2*capacitorEnergy/capacitance) ** 0.5
    #charge
    charge = capacitance * voltage
    #capacitor energy (subbed in)
    #capacitorEnergy = 0.5 * charge * voltage
    current = voltage/resistance
    #power
    power = voltage * current
    
    #kinematics
    velocity = (2*kineticEnergy/mass) ** 0.5
    acceleration = calcAcceleration()
    displacement += velocity * dt
    
    #sum of forces, drag, friction, motionResistance
    motionResistance = gravity * mass * coefficientOfRollingResistance / wheelRadius
    friction = 0
    drag = 0.5 * airDensity * (velocity**2) * crossSectionalArea * dragCoefficient
    forcesLoss = (motionResistance * 2) + friction + drag
    force = (acceleration * mass) - forcesLoss
    
    #mechanical energy losses
    mechanicalPowerLoss = forcesLoss * velocity
    kineticEnergy -= mechanicalPowerLoss * dt
    
    #add ICEngine generator power to charge
    kineticEnergy += generatorPower
    generatorEnergy += generatorPower * dt
    
    #interate
    update()
    t += dt
        
kmhrs = [vel*3.6 for vel in velocities]

plotThem()
importantInfo()

et = time.monotonic()
print("End time:", et - st)
```
